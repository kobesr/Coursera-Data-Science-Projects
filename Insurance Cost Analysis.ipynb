# **Insurance Cost Analysis**



In this project, I have performed analytics operations on an insurance database that uses the below mentioned parameters.

| Parameter |Description| Content type |
|---|----|---|
|age| Age in years| integer |
|gender| Male or Female|integer (1 or 2)|
| bmi | Body mass index | float |
|no_of_children| Number of children | integer|
|smoker| Whether smoker or not | integer (0 or 1)|
|region| Which US region - NW, NE, SW, SE | integer (1,2,3 or 4 respectively)| 
|charges| Annual Insurance charges in USD | float|



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
import requests
import pandas as pd
from io import StringIO

filepath = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/medical_insurance_dataset.csv'

response = requests.get(filepath)
response.raise_for_status()  # ensure download succeeded

df = pd.read_csv(StringIO(response.text), header=None)

print(df.head(10))
```

```python
headers = ["age", "gender", "bmi", "no_of_children", "smoker", "region", "charges"]
df.columns = headers
df.replace('?', np.nan, inplace = True)
print(df.head(10))
```

# Data Wrangling

Identifying the columns that have some 'Null' or 'NaN' information.

```python
print(df.info())
```

Handle missing data:

- For continuous attributes (e.g., age), replace missing values with the mean.
- For categorical attributes (e.g., smoker), replace missing values with the most frequent value.
- Update the data types of the respective columns.
- Verify the update using `df.info()`.


```python
# smoker is a categorical attribute, replace with most frequent entry
is_smoker = df['smoker'].value_counts().idxmax()
df["smoker"].replace(np.nan, is_smoker, inplace=True)

# age is a continuous variable, replace with mean age
mean_age = df['age'].astype('float').mean(axis=0)
df["age"].replace(np.nan, mean_age, inplace=True)

# Update data types
df[["age","smoker"]] = df[["age","smoker"]].astype("int")

print(df.info())
```

Updating the `charges` column to have all values rounded to 2 decimal places.

```python
df[["charges"]] = np.round(df[["charges"]],2)
print(df.head())
```

# Exploratory Data Analysis

Implementing a regression plot for `charges` with respect to `bmi` and a boxplot for `charges` with respect to `smoker`.

```python
sns.regplot(x="bmi", y="charges", data=df, line_kws={"color": "red"})
plt.ylim(0,)
```

```python
sns.boxplot(x="smoker", y="charges", data=df)
```

# Model Development
Fitting a regression model that may be used to predict the `charges` value just by using the `smoker` attribute of the dataset. The R^2 was then calculated.

```python
X = df[['smoker']]
Y = df['charges']
lm = LinearRegression()
lm.fit(X,Y)
print('The R^2 score is:',lm.score(X, Y))
```

Fitting a linear regression model that may be used to predict the `charges` value by using all other attributes of the dataset. R^2 was calculated.
Note: A model improvement was observed.

```python
# definition of Y and lm remain same as used in last cell. 
Z = df[["age", "gender", "bmi", "no_of_children", "smoker", "region"]]
lm.fit(Z,Y)
print('The improved R^2 score is:',lm.score(Z, Y))
```

A training pipeline was created that uses `StandardScaler()`, `PolynomialFeatures()` and `LinearRegression()` to create a model that can predict the `charges` value using all the other attributes of the dataset. 
Note: There should be even further improvement on the model performance.

```python
# Y and Z use the same values as defined in previous cells 
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe=Pipeline(Input)
Z = Z.astype(float)
pipe.fit(Z,Y)
ypipe=pipe.predict(Z)
print(r2_score(Y,ypipe))
```

# Model Refinement
The data was split into training and testing subsets, assuming that 20% of the data will be reserved for testing.

```python
x_train, x_test, y_train, y_test = train_test_split(Z, Y, test_size=0.2, random_state=1)
```

A Ridge regressor was initialized that uses a hyperparameter of $ \alpha = 0.1 $. The model was fitted using the training data subset and the R^2 score calculated for the testing data.

```python
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)
yhat = RidgeModel.predict(x_test)
print(r2_score(y_test, yhat))
```

A polynomial transformation to the training parameters with degree = 2 was applied. This transformed feature was set to fit the same regression model as above, using the training subset. The R^2 score was calculated for the testing subset.

```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
RidgeModel.fit(x_train_pr, y_train)
y_hat = RidgeModel.predict(x_test_pr)
print(r2_score(y_test, y_hat))
```
