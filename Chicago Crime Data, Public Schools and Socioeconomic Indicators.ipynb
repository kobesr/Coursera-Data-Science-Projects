### Store the datasets in database tables

To analyze the data using SQL, it first needs to be loaded into SQLite DB.
We will create three tables in as under:

1.  **CENSUS_DATA**
2.  **CHICAGO_PUBLIC_SCHOOLS**
3.  **CHICAGO_CRIME_DATA**


Loading the pandas and sqlite3 libraries and establishing a connection to FinalDB.db

```python
%load_ext sql
%sql sqlite:///FinalDB.db
```

Using Pandas to load the data available in the links above to dataframes. These dataframes will then be used to load data onto the database FinalDB.db as required tables. A connection between SQL maginc module and the database was established.

```python
import pandas
import sqlite3

con = sqlite3.connect("FinalDB.db")
cur = con.cursor()

#Chicago Census Data
df1 = pandas.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01")
df1.to_sql("Chicago_Census_Data", con, if_exists ='replace', index =False)

#Chicago Public Schools
df2 = pandas.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01")
df2.to_sql("Chicago_Public_Schools", con, if_exists = 'replace', index = False)

#Chicago Crime Data
df3 = pandas.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01")
df3.to_sql("Chicago_Crime_Data", con, if_exists = 'replace', index = False)
```

Finding the total number of crimes recorded in the CRIME table.

```python
%sql SELECT COUNT(*) AS Total_Crimes FROM Chicago_Crime_Data;
```

Listing community area names and numbers with per capita income less than 11,000.
Note: SQL magic format changed to better showcase SQL code structure.

```python
%%sql
SELECT COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER 
FROM Chicago_Census_Data 
WHERE PER_CAPITA_INCOME < 11000;
```

Listing all case numbers for crimes involving minors. (Children are not considered minors for the purposes of crime analysis)

```python
%%sql
SELECT CASE_NUMBER 
FROM Chicago_Crime_Data 
WHERE DESCRIPTION LIKE '%MINOR%';
```

Listing all kidnapping crimes involving a child.

```python
%%sql
SELECT *
FROM Chicago_Crime_Data 
WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%';
```

Listing the kind of crimes that were recorded at schools. (No repetitions)

```python
%%sql
SELECT DISTINCT PRIMARY_TYPE
FROM Chicago_Crime_Data 
WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%';
```

Lisitng the type of schools along with the average safety school for each type (ES for Elementary School, MS for Middle School and HS for High School)

```python
%%sql 
SELECT `Elementary, Middle, or High School` AS School_Type,AVG(SAFETY_SCORE) AS avg_Safety_Score 
FROM Chicago_Public_Schools 
WHERE School_Type IN ('ES','MS','HS') 
GROUP BY School_Type
```

Listing 5 community areas with highest % of households below poverty line.

```python
%%sql
SELECT COMMUNITY_AREA_NAME
FROM Chicago_Census_Data
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
LIMIT 5;
```

Checking which community area is most crime prone and displaying the community area number only.

```python
%%sql
SELECT COUNT(*) AS Total_Crimes, COMMUNITY_AREA_NUMBER
FROM Chicago_Crime_Data
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY Total_Crimes DESC
LIMIT 1;
```

Using a sub-query to find the name of the community area with highest hardship index.

```python
%%sql
SELECT COMMUNITY_AREA_NAME
FROM Chicago_Census_Data
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM Chicago_Census_Data);
```

Using a sub-query to determine the Community Area Name with most number of crimes.

```python
%%sql
SELECT COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME
FROM Chicago_Census_Data
WHERE COMMUNITY_AREA_NUMBER = (
    SELECT COMMUNITY_AREA_NUMBER
    FROM Chicago_Crime_Data
    GROUP BY COMMUNITY_AREA_NUMBER
    ORDER BY COUNT(*) DESC
    LIMIT 1
);
```
